```python
import pandas as pd

df = pd.read_csv("marketing_campaign.csv", sep="\t")

print(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Income"] = df["Income"].fillna(df["Income"].median())
```

```python
df.drop_duplicates(subset ='ID')
```

```python
df["Marital_Status"] = df["Marital_Status"].str.strip()
df["Education"] = df["Education"].str.strip()
```

```python
df["Marital_Status"] = df["Marital_Status"].str.title()
df["Education"] = df["Education"].str.title()
```

```python
df.head()
```

```python
(df["Education"].unique())
```

```python
df["Education"] = df["Education"].replace({
    "phd":"PhD",
    "2N Cycle":"2nd Cycle"
})
```

```python
(df["Marital_Status"].unique())
```

```python
df["Marital_Status"] = df["Marital_Status"].replace({
    "Alone":"Single",
    "Absurd":"Other",
    "Yolo":"Other"
})
```

```python
print(df["Marital_Status"].value_counts())
print(df["Education"].value_counts())
```

```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], dayfirst=True)
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ", "_")
```

```python
df = df.rename(columns={
    "year_birth":"age",
    "mntwines": "wine_spending",
    "mntfruits": "fruit_spending",
    "mntmeatproducts": "meat_spending",
    "dt_customer": "customer_join_date",
    "mntfishproduct": "amount_spent_fish",
    "numdealspurchases":"num_discount_purchases",
    'mntsweetproducts': 'amount_spent_sweets',
    'mntgoldprods': 'amount_spent_gold',
     'numwebpurchases': 'num_online_purchases',
    'numcatalogpurchases': 'num_catalog_purchases',
    'numstorepurchases': 'num_store_purchases',
    'numwebvisitsmonth': 'num_website_visits_month',
    'acceptedcmp3': 'accepted_campaign_3',
    'acceptedcmp4': 'accepted_campaign_4',
    'acceptedcmp5': 'accepted_campaign_5',
    'acceptedcmp1': 'accepted_campaign_1',
    'acceptedcmp2': 'accepted_campaign_2',
    'complain': 'has_complained',
    'z_costcontact': 'cost_to_contact',
    'z_revenue': 'company_revenue',
    'response': 'campaign_response'
})
```

```python
df.columns
```

```python
df.dtypes
```

```python
df["age"] =2025 - df["age"] 
```

```python
df["age"]= df["age"].astype(int)
```

```python
df["income"] = pd.to_numeric(df["income"], errors="coerce")
```

```python
df["education"] = df["education"].astype("category")
df["marital_status"] = df["marital_status"].astype("category")

```

```python
import matplotlib.pyplot as plt

plt.boxplot(df["income"])
plt.title("Boxplot of Income")
plt.show()
```

```python
num_cols = ['income', 'kidhome',
       'teenhome', 'customer_join_date', 'recency', 'wine_spending',
       'fruit_spending', 'meat_spending', 'mntfishproducts',
       'amount_spent_sweets', 'amount_spent_gold', 'num_discount_purchases',
       'num_online_purchases', 'num_catalog_purchases', 'num_store_purchases',
       'num_website_visits_month', 'accepted_campaign_3',
       'accepted_campaign_4', 'accepted_campaign_5', 'accepted_campaign_1',
       'accepted_campaign_2', 'has_complained', 'cost_to_contact',
       'company_revenue', 'campaign_response']

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python

df = df[(df["age"] > 0) & (df["age"] < 100)]
```

```python
df = df.drop(columns=["cost_to_contact", "company_revenue"])
```

```python
df["total_spent"] = (
    df["wine_spending"] +
    df["fruit_spending"] +
    df["meat_spending"] +
    df["mntfishproducts"] +
    df["amount_spent_sweets"] +
    df["amount_spent_gold"]
)
```

```python
df["total_purchases"] = (
    df["num_discount_purchases"] +
    df["num_online_purchases"] +
    df["num_catalog_purchases"] +
    df["num_store_purchases"]
)
```

```python
df["customer_days"] = (pd.Timestamp.today() - df["customer_join_date"]).dt.days
```

```python
df.to_csv("final_marketing_cleaned.csv", index=False)
```
